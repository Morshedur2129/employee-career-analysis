```python
import pandas as pd
import numpy as np
import plotly.express
import plotly.graph_objects as go
```

```python
data=pd.read_csv("/content/synthetic_survey_dataset_500.csv")
```

```python
data.head()
```

```python
country= data["Country"].value_counts()
label=country.index
counts=country.values
color=['red','lightgreen']
fig=go.Figure(data=[go.Pie(labels =label,values=counts)])
fig.update_layout(title_text="Survey Results")
fig.update_traces(hoverinfo='label+value', textinfo='percent', textfont_size=30,
                  marker=dict(colors=color, line=dict(color='black', width=3)))
fig.show()
```

```python
question3="Preferred Work Setup"
question3=data[question3].value_counts()
label1=question3.index
counts1=question3.values
colors=['red','lightgreen']
fig=go.Figure(data=[go.Pie(labels=label1,values=counts1)])
fig.show()
```

```python

teamwork_data = data[~data['Preferred Work Setup'].str.contains('Work alone')]

teamwork_by_country = teamwork_data['Country'].value_counts().reset_index()
teamwork_by_country.columns = ['Country', 'Teamwork Count']

display(teamwork_by_country)
```

The relationship between 'Gender' and 'Preferred Work Setup':

```python
# Filter data for 'Female' gender
women_data = data[data['Gender'] == 'Female']

# Create a cross-tabulation of 'Country' and 'Preferred Work Setup' for women
women_team_preference_by_country = pd.crosstab(women_data['Country'], women_data['Preferred Work Setup'])

# Display the table
display(women_team_preference_by_country)
```
